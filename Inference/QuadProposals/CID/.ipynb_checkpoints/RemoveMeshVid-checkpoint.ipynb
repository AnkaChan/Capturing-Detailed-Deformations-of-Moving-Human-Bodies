```python
f = open('CID_list.txt', 'r')
wf = open('CID_no_meshVID.txt', 'w')
wfCIDToMeshVID = open('CIDToMeshVID.txt')
wfTetMeshIdToCID = open('MeshVIDToCID.txt', 'w')
```

```python
import numpy as np
```

```python
CIDListNoVIDs = []
CIDToMeshVID = []
MeshVIDToCIDList =  -np.ones((23589,1), dtype=np.int32)
cid = 0
for line in f.readlines():
    tokens = line.split(' ')
    CIDListNoVIDs.append(tokens[0:-1])
    CIDToMeshVID.append(tokens[-1])
    MVID = int(tokens[-1])
    if MVID != -1:
        #print(line, tokens)
        print(MVID)
        MeshVIDToCIDList[MVID-1] = cid
    cid = cid + 1
    
CIDListNoVIDs    

```

```python
for line in CIDToMeshVID:
    wfCIDToMeshVID.write(line + '\n')
wfCIDToMeshVID.close()
```

```python
for line in CIDListNoVIDs:
    wf.write(' '.join(line) + '\n')
```

```python
for i in range(MeshVIDToCIDList.shape[0]):
    wfTetMeshIdToCID.write(str(MeshVIDToCIDList[i,0]) + '\n')

  
```
